```python
def split_and_flatten(row):
    impressions = row['impression'].split()
    flattened = [item.split('-')[0] for item in impressions]
    return ' '.join(flattened)


df['new_dataset'] = df.apply(split_and_flatten, axis=1)


print(df[['new_dataset']])
```

```python
new_df = df['new_dataset'].str.split().explode().reset_index(drop=True)

new_df = new_df.rename('split_values')


df6 = pd.DataFrame({'auction_id': new_df})


print(df6)
```

```python
df5=pd.read_csv('df11 (3).csv')
df5
```

```python
df = pd.concat([df6, df5], ignore_index=True)

df
```
